## 信息熵

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)

def gini(p):
    return 2 * p * (1 - p)

def err(p):
    return 1- np.max(np.vstack((p, 1-p)), 0)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x), 'b-', label='entory')
plt.plot(x, gini(x), 'r-', label='gini')
plt.plot(x, err(x), 'g-', label='error')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
